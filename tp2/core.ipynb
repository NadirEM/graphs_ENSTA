```python
import pandas as pd 
degrees = pd.read_csv('c_degrees.csv', sep = ' ', header = None)[[0,1]]
cores = pd.read_csv('cores.csv', sep = ' ', header = None)[[0,1]]
```

```python
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
```

```python
rho = pearsonr(list(cores[1]), list(degrees[1]))
rho[0]
```

```python
fig= plt.figure()

ax=fig.add_axes([0,0,1,1])
#l = list(dict.values())
x = list(degrees[1])
y = list(cores[1])
ax.scatter(x, y)
ax.set_ylabel('Degree', fontsize=25)
ax.set_xlabel("Coerness", fontsize=25)
plt.plot(range(14), range(14))
plt.show()
```
