# **Import**

```
import networkx as nx 
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics.cluster import normalized_mutual_info_score, adjusted_rand_score
```

# **Visualizations of graphs and Propagation algorithm partitions**

```
G=nx.read_edgelist("graph_0.8_0.4.txt")
nx.draw(G) 
plt.draw()  
plt.savefig('graph_0.8_0.4.png')
```

```
node_labels = {}
with open("comm_graph_0.8_0.4.txt") as f:
    for line in f:
       (key, val) = line.split()
       node_labels[key] = int(val)
default_colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']
x = np.unique(list(node_labels.values()))
default_colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf'][:len(x)]
colors = []
for node in G.nodes():
  idx, = np.where(x == node_labels[str(node)])[0]
  colors.append(default_colors[idx])

nx.draw_networkx(G, pos=None, with_labels=False, labels=node_labels, node_color=colors)
plt.draw()  
plt.savefig('comm_graph_0.8_0.4.png')
```

# **Evaluation**

```
labels_true = np.array(list(np.asarray(np.zeros(100))) + list(np.asarray(np.ones(100))) + list(2*np.asarray(np.ones(100))) + list(3*np.asarray(np.ones(100))))

labels_pred_propagation_008 = np.loadtxt('comm_graph_0.8_0.08.txt')[:,1]  # ratio = 10
labels_pred_louvain_008 = np.loadtxt('comm_louvain_0.08.txt')[:,1]  # ratio = 10
labels_pred_propagation_005 = np.loadtxt('comm_graph_0.8_0.05.txt')[:,1] # ratio = 16
labels_pred_louvain_005 = np.loadtxt('comm_louvain_0.05.txt')[:,1]  # ratio = 16

print('Results for the random graph with ratio = 10 :\n');

print('The Normalized Mutual Information for the label propagation algorithm is : ', round(normalized_mutual_info_score(labels_true, labels_pred_propagation_008),3))
print('The Adjusted Rand index for the label propagation algorithm is : ', round(adjusted_rand_score(labels_true, labels_pred_propagation_008),3))

print('The Normalized Mutual Information for the Louvain algorithm is : ', round(normalized_mutual_info_score(labels_true, labels_pred_louvain_008),3))
print('The Adjusted Rand index for the Louvain algorithm is : ', round(adjusted_rand_score(labels_true, labels_pred_louvain_008),3))


print('\nResults for the random graph with ratio = 16 :\n');

print('The Normalized Mutual Information for the label propagation algorithm is : ', round(normalized_mutual_info_score(labels_true, labels_pred_propagation_005),3))
print('The Adjusted Rand index for the label propagation algorithm is : ', round(adjusted_rand_score(labels_true, labels_pred_propagation_005),3))

print('The Normalized Mutual Information for the Louvain algorithm is : ', round(normalized_mutual_info_score(labels_true, labels_pred_louvain_005),3))
print('The Adjusted Rand index for the Louvain algorithm is : ', round(adjusted_rand_score(labels_true, labels_pred_louvain_005),3))

```
